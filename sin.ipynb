```python
import numpy as np
import math
import matplotlib.pyplot as plt
import matplotlib
matplotlib.style.use('ggplot')
fig_size = plt.rcParams["figure.figsize"]
fig_size[0] = 20
fig_size[1] = 10
plt.rcParams["figure.figsize"] = fig_size
```

```python
x = np.linspace(0, 2*np.pi)

y_sum = np.zeros(50)
for i in range(5):
    y = np.sin(x * (i+1))
    y_sum += y
plt.plot(x, y_sum)
plt.show()
```

```python
fi = np.random.randint(0, 2*np.pi, size = 50)
fi

fi2 = np.random.random_sample(5,) * 10
fi2

fi3 = [1.5, 1.2, 1.3, 1.4, 1.6]
```

При значениях около 1.5 достигаются максимальные значения

```python
for fij in fi3:
    y_sum = np.zeros(50)
    for i in range(500):
        y = np.sin(x * (i+1) + fij)
        y_sum += y
    plt.plot(x, y_sum)
    plt.show()
```
