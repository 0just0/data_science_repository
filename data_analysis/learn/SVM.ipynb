# Машина опорных векторов (Support Vector Machine, SVM) и ядерные методы

**Метод опорных векторов (англ. SVM, support vector machine)** — набор схожих алгоритмов обучения с учителем, использующихся для задач классификации и регрессионного анализа. 

Сначала рассмотрим интуитивное объяснение алгоритма SVM, а затем более формальное.

## Что такое SVM (Support Vector Machine)?

Машина опорных векторов (support vector machine, SVM) - алгоритм машинного обучения с учителем, используемый как для решения задач классификации, так и для решения задач регрессии.

Однако, SVM зачастую используется в задачах классификации. 

Итак, как обычно, каждое наблюдение рассматривается как точка в $n$-мерном пространстве ($n$ - число рассматриваемых атрибутов) со значениями каждого атрибута как значение по соответствующей координате.

Затем, выполняется классификации при помощи поиска гиперплоскости, которая способна достаточно хорошо различить два класса между собой.

![](https://www.analyticsvidhya.com/wp-content/uploads/2015/10/SVM_1.png)

Опорные векторы - просто являются координатами отдельных наблюдений тренировочной выборки. SVM современным алгоритмом, который достаточно хорошо разделяет два класса.

## Как работает SVM (интуитивный уровень)?

Итак, выше мы сказали о том, что классы отделяются друг от друга с помощью гиперплоскости. 

Разумный вопрос: как идентифицировать правильную гиперплоскость?

#### Поиск оптимальной гиперплоскости (сценарий 1):

Пусть мы имеем три гиперплоскости (A, B и C). Найдем правильную гиперплоскость, чтобы идентифицировать звёзды и круги.
![](https://www.analyticsvidhya.com/wp-content/uploads/2015/10/SVM_21.png)

Следует запомнить хорошее эмпирическое правило для нахождения верной гиперплоскости:
*Выбирайте гиперплоскость, которая лучше разделяет два класса.*

В приведенном сценарии гиперплоскость "B" выполняет эту задачу лучше остальных.


#### Поиск оптимальной гиперплоскости (сценарий 2):

Пусть мы имеем три гиперплоскости (A, B и C) и все они хорошо разделяют классы. Как сейчас идентифицировать корректную гиперплоскость?

![](https://www.analyticsvidhya.com/wp-content/uploads/2015/10/SVM_3.png)

В этом случае, максимизация расстояния между ближайшими точками (каждого класса) и гиперплоскостью поможет нам решить, какая же гиперплоскость является лучше. Это расстояние называется "зазор" (margin). 

![](https://www.analyticsvidhya.com/wp-content/uploads/2015/10/SVM_4.png)

Итак, можно увидеть, что зазор (margin) для гиперплоскости C больше, по сравнению с A и B. Поэтому, мы выбираем гиперплоскость C в качестве оптимальной.

Еше одной причиной для выбора гиперплоскости с наибольшим зазором (higher margin) является устойчивость (робастностьб robastness) модели. При выборе гиперплоскость с небольшим зазором будет достаточно большой риск неправильной классификации.


#### Поиск оптимальной гиперплоскости (сценарий 3):

*Подсказка: используем правила, которые обсуждались выше для поиска правильной гиперплоскости*

![](https://www.analyticsvidhya.com/wp-content/uploads/2015/10/SVM_5.png)

Возможно, выбрана гиперплоскость $B$, так как она имеет больший зазор (margin) по сравнению с гиперплоскостью $A$. 

Однако, здесь есть подвох: SVM выбирает гиперплоскость, которая прежде всего максимизирует точность классификации, и лишь затем максимизирует зазор. Соответственно, гиперплоскость B будет иметь ошибку классификации, а A будет проводить корректную классификацию.

Итак, верной гиперплоскостью в этом случае является $A$.




#### Поиск оптимальной гиперплоскости (сценарий 4):

На иллюстрации ниже нельзя провести разделение на два класса с использованием прямой линии, так как один объект (звезда) располагается в области другого класса (кругов), и является выбросом (outlier).

![](https://www.analyticsvidhya.com/wp-content/uploads/2015/10/SVM_61.png)

Как уже упомянуто, один объект (класса звезда) по сравнению с другими является выбросом (outlier). SVM имеет особенность, позволяющую ему игнорировать выбросы и находить гиперплоскость, имеющую максимальный зазор.

Поэтому можно сказать, что SVM устойчив к выбросам. 

![](https://www.analyticsvidhya.com/wp-content/uploads/2015/10/SVM_71.png)

#### Поиск оптимальной гиперплоскости (сценарий 5):

В этом сценарии у нас нет линейной гиперплоскости между двумя классами. Как SVM будет классифицировать эти два класса?

До сих пор мы рассматривали только линейные гиперплоскости.

![](https://www.analyticsvidhya.com/wp-content/uploads/2015/10/SVM_8.png)

SVM может решить эту проблему! SVM решает эту задачу вводя дополнительную особенность. Мы добавляем новый атрибут $z = x^2 + y^2$. Сейчас, построим точки с использованием осей $x$ и $z$:

![](https://www.analyticsvidhya.com/wp-content/uploads/2015/10/SVM_9.png)

На графике выше все значения $z$ будут положительными, поскольку $z$ является суммой квадратов $x$ и $y$;

В алгоритме SVM легко получить линейную гиперплоскость между двумя такими классами. Но здесь возникает вопросом - должны ли мы вручную добавлять новые атрибуты, чтобы получить гиперплоскость?

Нет, SVM имеет технологию, называемую **ядерным методом (kernel trick)**. Они являются функциями, которые берут входное пространство атрибутов малой размерности и преобразют его в многомерное пространство, т.е. преобразуют неразделимую задачу, в разделимую, где функции называются **ядрами** (kernels). Наиболее полезным использованием их является использование в задачах, не являющихся линейно-разделимыми.

Говоря проще, SVM выполняет ряд достаточно сложных преобразований данных и находит процесс, в котором данные являются разделимыми относительно целевого класса.

## Использование SVM в scikit-learn

Библиотека scikit-learn также содержит функции, позволяющие работать с алгоритмом SVM.

Рассмотрим участок кода, работающий с SVM


```python
#Import Library
from sklearn.svm import SVC
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split

# Рассмотрим задачу классификации цветов Ириса Фишера
iris = load_digits()
X = iris.data
y = iris.target

X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.33, random_state=42)

# Create SVM classification object 
model = SVC(kernel='linear', C=1, gamma=1) 

# there is various option associated with it, like changing kernel, gamma and C value. Will discuss more 
model.fit(X_train, y_train)

print('Score: {0}'.format(model.score(X_test, y_test)))
# Method score() - eturns the mean accuracy on the given test data and labels.
```

### Параметр kernel

**kernel**: в scikit-learn имеются разные варианты используемых ядер, например *linear*, *rbf*, *poly* и некоторые другие (параметром по умолчанию является использование ядра типа *rbg*).

*rbf* и *poly* полезные для нелинейной гиперплоскости. Рассмотрим пример, где использовано линейное ядро для двух атрибутов из набора данных - Ирисы Фишера для классификации.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm, datasets

# import some data to play with
iris = datasets.load_iris()
X = iris.data[:, :2] 
y = iris.target
```

#### Линейное ядро

```python
# Создаем объект класса SVM и обучаем на данных (не разделяем поскольку хотим посмотреть на разрешающую границу)
C = 1.0 # SVM regularization parameter
svc = svm.SVC(kernel='linear', C=1).fit(X, y)
```

```python
# create a mesh to plot in
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
h = (x_max / x_min)/100
xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
 np.arange(y_min, y_max, h))
plt.subplot(1, 1, 1)
Z = svc.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)
plt.contourf(xx, yy, Z, cmap=plt.cm.Paired, alpha=0.8)
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Paired)
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')
plt.xlim(xx.min(), xx.max())
plt.title('SVC with linear kernel')
plt.show()
```

#### Ядро RBF

```python
svc = svm.SVC(kernel='rbf', C=1,gamma='auto').fit(X, y)
# create a mesh to plot in
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
h = (x_max / x_min)/100
xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
 np.arange(y_min, y_max, h))
plt.subplot(1, 1, 1)
Z = svc.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)
plt.contourf(xx, yy, Z, cmap=plt.cm.Paired, alpha=0.8)
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Paired)
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')
plt.xlim(xx.min(), xx.max())
plt.title('SVC with RBF kernel')
plt.show()
```

Можно использовать линейное ядро в том случае, если имеется большое число атрибутов (например, более 1000), поскольку более вероятно, что данные будут линейно разделимы в пространстве высокой размерности.

Кроме того, при использовании ядра RBF не забывайте использовать кросс-валидацию для выбранных параметров для того, чтобы избежать переобучения.

### Параметр gamma

**Параметр gamma** соответствует коэффициенту сл. ядер: RBF, полиномиальное и сигмоидальное ядро.
Чем выше значение $\gamma$, тем более точно алгоритм будет пытаться подстроиться под тренировочную выборку, что может вызвать переобучение.

В качестве примера рассмотрим различные коэффициенты $\gamma$: $0, 10$ и $100$.

```python
svc = svm.SVC(kernel='rbf', C=1,gamma='auto').fit(X, y)
# create a mesh to plot in
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
h = (x_max / x_min)/100
xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
 np.arange(y_min, y_max, h))
plt.subplot(1, 1, 1)
Z = svc.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)
plt.contourf(xx, yy, Z, cmap=plt.cm.Paired, alpha=0.8)
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Paired)
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')
plt.xlim(xx.min(), xx.max())
plt.title('SVC with RBF kernel, gamma = auto ({0})'.format(float(1.0/svc.support_vectors_.shape[1])))
plt.show()

svc = svm.SVC(kernel='rbf', C=1,gamma=1).fit(X, y)
# create a mesh to plot in
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
h = (x_max / x_min)/100
xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
 np.arange(y_min, y_max, h))
plt.subplot(1, 1, 1)
Z = svc.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)
plt.contourf(xx, yy, Z, cmap=plt.cm.Paired, alpha=0.8)
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Paired)
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')
plt.xlim(xx.min(), xx.max())
plt.title('SVC with RBF kernel, gamma = 1')
plt.show()


svc = svm.SVC(kernel='rbf', C=1,gamma=10).fit(X, y)
# create a mesh to plot in
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
h = (x_max / x_min)/100
xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
 np.arange(y_min, y_max, h))
plt.subplot(1, 1, 1)
Z = svc.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)
plt.contourf(xx, yy, Z, cmap=plt.cm.Paired, alpha=0.8)
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Paired)
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')
plt.xlim(xx.min(), xx.max())
plt.title('SVC with RBF kernel, gamma = 10')
plt.show()

svc = svm.SVC(kernel='rbf', C=1,gamma=100).fit(X, y)
# create a mesh to plot in
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
h = (x_max / x_min)/100
xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
 np.arange(y_min, y_max, h))
plt.subplot(1, 1, 1)
Z = svc.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)
plt.contourf(xx, yy, Z, cmap=plt.cm.Paired, alpha=0.8)
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Paired)
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')
plt.xlim(xx.min(), xx.max())
plt.title('SVC with RBF kernel, gamma = 100')
plt.show()

```

### Достоинства и недостатки SVM

**Достоинства**
* SVM достаточно хорошо работает при условии разделения данных;
* SVM эффективен в пространствах высокой размерности;
* SVM использует подмножество обучающей выборке при построении разрешающей функции (они называются опорными векторами), а также эффективен с точки зрения использования памяти.

**Недостатки**
* Когда мы имеем очень большие объемы данных, время обучения SVM может быть досточно долгим
* SVM может недостаточно хорошо работать, когда данные сильно зашумлены, т.е. целевые классы "перекрываются"
* SVM напрямую не дает вероятностные оценки, и они могут быть вычислены с использованием крайне трудоемкой операции кросс-валидации.

# SVM : Формальная постановка задачи

### Понятие гиперплоскости

В $p$-мерном пространстве, **гиперплоскосью** называется плоское аффинное подпространство, размерности $p-1$.

Пример:
* для двумерного пространства (плоскость) - гиперплоскостью является прямая
* для трехмерного пространства - плоскость.

Математическое определение гиперплоскости достаточно простое.

В двумерном пространстве гиперплоскость с параметрами $\beta_0$, $\beta_1$, $\beta_2$ задается уравнением:
$$\beta_0 + \beta_1 x_1 + \beta_2 x_2 = 0$$

Это уравнение легко обобщить на $p$-мерное проостранство:
$$\beta_0 + \beta_1 x_1 + \beta_2 x_2 + \ldots + \beta_p x_p = 0$$

Это уравнение задает $p$-мерную гиперплоскость в том смысле, что любая точка $x = (x_1, x_2, \ldots, x_p)^T$ в $p$-мерном пространстве удовлетворяющая этому уравнению лежит на гиперплоскости.

Пусть теперь $x$ не удовлетворяет этому условию. 

Тогда, может выполниться, например, сл. условие:
$$\beta_0 + \beta_1 x_1 + \beta_2 x_2 + \ldots + \beta_p x_p < 0$$

Это говорит о том, что $x$ назодится на определенной стороне относительно гиперплоскости. 

А если: 
$$\beta_0 + \beta_1 x_1 + \beta_2 x_2 + \ldots + \beta_p x_p > 0$$

то $x$ находится по другую сторону о гиперплоскости. Таким образом, можно думать о гиперплоскости как о чем-то, что разделяет $p$-мерное проостранство на две половины.

### Классификация с использованием гиперплоскости

Представим, что у нас есть матрица $X$ размером $p\times n$ ($p$ - атрибутов, $n$ - наблюдений).

$$x_1 = (x_{11}, x_{12}, \ldots, x_{1p})$$
$$x_2 = (x_{21}, x_{22}, \ldots, x_{2p})$$
$$\ldots$$
$$x_n = (x_{n1}, x_{n2}, \ldots, x_{np})$$

И эти наблюдения относятся к двум классам, т.е. $y_1, \ldots, y_n \in \{-1, 1\}$.

Кроме того, у нас есть контрольное наблюдение: $x^{*} = (x_1^{*}, \ldots, x_p^{*})$. Нашей целью является создание классификатора на основе обучающих данных, который позволит правильно предсказывать контрольное наблюдение по значениям его признаков.

Мы уже изучили ряд подходов для решения этой задачи, а теперь изучим подход, основанный на понятии *разделяющей гиперплоскости*. 

Представим, что имеется возможность сконструировать такую гиперплоскость, которая разделяет обучающие наблюдения в точном соответствии с метками их классов. На рис. 2 (слева) можно увидеть примеры подобных гиперплоскостей.
![](img/svm/Theory2.png "Рис. 2")

Наблюдения "синего класса", обозначим за $y_i = 1$, а "красного класса" за $y_i = -1$.

Тогда разделяющая плоскость будет иметь сл. свойства:
$$\beta_0 + \beta_1 x_{i1} + \beta_2 x_{i2} + \ldots + \beta_p x_{ip} > 0 \mbox{  если  } y_i =1$$
и
$$\beta_0 + \beta_1 x_{i1} + \beta_2 x_{i2} + \ldots + \beta_p x_{ip} < 0 \mbox{  если  } y_i =-1$$

Можно также сказать, что разделяющей является гиперплоскость со сл. свойством: 
$$y_i (\beta_0 + \beta_1 x_{i1} + \beta_2 x_{i2} + \ldots + \beta_p x_{ip}) > 0 \mbox{ для всех } i = 1,\ldots,n$$


Если разделяющая плоскость существует, то её можно использовать для построения естественного классификатора: контрольное наблюдение классифицируют в зависимости от того, по какую сторону от гиперплоскости оно находится.

На рис. 2 справа приведен пример такого классификатора.

Иными словами, мы классифицируем контрольное наблюдение $x^{*}$ в зависимости от знака функции $f(x^{*}) = \beta_0 + \beta_1 x_1^{*} + \beta_2 x_2^{*} + \ldots + \beta_p x_p^{*}$.

Если $f(x^{*})>0$, то контрольное наблюдение относим к классу $1$, если $f(x^{*})<0$, то к классу $-1$. Можно также принять во внимание **конкретное значение $f(x^{*})$** и воспринимать его как степень уверенности в правильной классификации.

Чем ближе значение $f(x^{*})$ к нулю, тем меньше уверенность в правильной классификации, и наоборот.

**Классификатор онованный на разделяющей гиперплоскости приводит к линейной решающей границе.**

### Классификатор с максимальным зазором

Если наши данные можно *идеально* разделить с помощью гиперплоскости, то таких гиперплоскостей (обычно) будет бесконечно много. Это обусловлено тем, что некоторую гиперплоскость можно сдвинуть на очень небольшое расстояние вверх или вниз, не задев ни одно из наблюдений.
![](img/svm/Theory2.png "Рис. 2")

На рис. 2 слева показаны три возможные гиперплоскости. Для создания классификатора на основе разделяющей гиперплоскости необходим какой-либо приемлемый способ принятия решения в отношении того, какую их бесконечно большого числа возможных гиперплоскостей выборать.

Естественным выбором является **гиперплоскость с максимальным зазором** (также *оптимальная разделяющая гиперплоскость*), представляющая собой *гиперплоскость, максимально удаленную от обучающих наблюдений*. Другими словами, можно вычислить (перпендикулярное) расстояние от каждого наблюдения до некоторой разделяющей гиперплоскости; минимальное из этих расстояний называют **зазором**. Гиперплоскость с максимальным зазлолм представляет собой разделяющую гиперплоскость, которая удалена от ближайших к ней обюучающих наблюдений на наибольшее расстояние.  Далее, можно классифицировать некоторое контрольное наблюдение с учетом того, по какую сторону ль гиперплоскости с максимальным зазором оно находится. Такой классификатор называется **классификатором с максимальным зазором**. 

Мы надеемся, что классификатор, обладающий максимальным зазором для обучающих данных будет иметь широкий зазор и для контрольных данных и, следовательно, будем корректно классифицировать контрольные наблюдения. Хотя классификатор с максимальным зазором часто дает хорошие результаты, при больших $p$ он может приводить к переобучению.

Если $\beta_0, \beta_1, \ldots, \beta_p$ - коэффициенты гиперплоскости с максимальным зазором, тогда классификатор с максимальным зазором отнесет некоторое контрольное наблюдение $x^{*}$ к тому или иному классу в заивисимости от знака  $f(x^{*}) = \beta_0 + \beta_1 x_1^{*} + \beta_2 x_2^{*} + \ldots + \beta_p x_p^{*}$.

На рис. 3 показана гиперплоскость, обеспечивающая максимальный зазор для данных,изображенных на рис. 2.
![](img/svm/Theory3.png "Рис. 3.")

Можно провест своеобразную аналогию: эта гиепрплоскость (изображенная на рис. 3) представляет собой срединную линию наиболее широкого "бруска", который можно вставить между двумя классами.


На рис. 3 можно увидеть, что 3 обучающих наблюдения находятся на одинаковом удалении от гиперплоскости с максимальным зазором и лежат на прерывистых линиях, обозначающих ширину зазора. Эти три наблюдения известны как **опорные векторы**, поскольку они являются векторами в $p$-мерном пространстве и "поддерживают" гиперплоскость в том смысле, что если эти точки незначительно сдвинуть, то сдвинется и сама гиперплоскость.

**Гиперплоскость с максимальным зазором зависит только от опорных векторов, но не остальных наблюдений: сдвиг любого из остальныхз наблюдений нисколько бы не повлиял на гиперплоскость, если только в результате того сдвига точка не пересечет границу зазора.**
Гиперплоскость с максимальным зазором зависит только от небольшого подмножества наблюдений, что является **очень важным свойством**.



### Построение классификатора с максимальным зазором

Теперь рассмотрим задачу нахождения гиперплоскости с максимальным зазором по выбороке из $n$ обучающих наблюдений $x_1, x_2, \ldots, x_n \in R^{p}$ и соответствующих им меток классов $y_1, \ldots, y_n \in \{-1,1\}$.

Такая гиперплоскость представляет собой решение следующей оптимизационной проблемы:

$$ max_{\beta_0, \beta_1, \ldots, \beta_p} M$$
$$ \mbox{ при условии, что } \sum_{j=1}^{p}\beta_j^{2}=1,$$
$$ y_i (\beta_0 + \beta_1 x_{i1} + \beta_2 x_{i2} + \ldots + \beta_{p} x_{ip}) \geq M, \forall i =1,\ldots, n.$$

В действительности, эта оптимизационная проблема проще, чем она выглядит.
Прежде всего, ограничение 
$$ y_i (\beta_0 + \beta_1 x_{i1} + \beta_2 x_{i2} + \ldots + \beta_{p} x_{ip}) \geq M, \forall i =1,\ldots, n.$$
гарантирует, что каждое наблюденрие будет лежать по правильную сторону от гиперплоскости при условии, что $M$ принимает положительное значение. 

(На самом деле, для этого нам достаточно было бы иметь $ y_i (\beta_0 + \beta_1 x_{i1} + \beta_2 x_{i2} + \ldots + \beta_{p} x_{ip}) > 0$,  таким образом, это ограничение было сделано с некоторым запасом, т.е. требуется, чтобы значение $M$ было положительным).

Во-вторых, ограничение:
$\sum_{j=1}^{p}\beta_j^{2}=1$ не является строгим ограничением, поскольку если гиперплоскость задается выражением $\beta_0 + \beta_1 x_{i1} + \beta_2 x_{i2} + \ldots + \beta_{p} x_{ip} = 0$, то с таким же успехом задается и выражением $k(\beta_0 + \beta_1 x_{i1} + \beta_2 x_{i2} + \ldots + \beta_{p} x_{ip})=0$ для любых $k \neq 0$.
Однако, $\sum_{j=1}^{p}\beta_j^{2}=1$ наполняет смыслом выражение 
$$ y_i (\beta_0 + \beta_1 x_{i1} + \beta_2 x_{i2} + \ldots + \beta_{p} x_{ip}) \geq M, \forall i =1,\ldots, n.$$

можно показать, что в присутствии этого ограничения перпендикулярное расстояние от $i$-го наблюдения до гиперплоскости определяется как $$y_i(\beta_0 + \beta_1 x_{i1} + \beta_2 x_{i2} + \ldots + \beta_p x_{ip}).$$

Поэтому эти ограничения позволяют убедиться, что каждое наблюдение лежит по правильную сторону от гиперплоскости, и как минимум на расстоянии $M$ от неё. Таким образом, $M$ представляет собой зазор нашей гиперплосккости и приведенная выше оптимизационная задача заключается в нахождении $\beta_0, \beta_1, \ldots, \beta_p$, которые максимизируют $M$.

А это, в свл. очередь, в точности является определением гиперплоскости с максимальным зазором.

При таком сценарии оптимизационная задача рассмотренная ранее для $M>0$ не имеет решения.

Обобщение классификатора с максимальным зазором на случай неразделимых классов дает метод, известный как **классификатор на опорных векторах.**

Классификатор, основанный на разделяющей гиперплоскости, который безупречно классифирует объекты тренировочной выборки может привести к чрезмерной чувствительности к отдельным наблюдениям. Это является проблемой, поскольку расстояние от наблюдения до гиперплоскости может быть рассмотрено как мера уверенности в том, что наблюдение корректно классифицировано. Более того, тот факт, что гиперплоскость с максимальным зазором чрезмерно чувствительна к изменениям в одном наблюдении допускает, что мы сильно подвержены переобучению.

В этом случае, мы можем рассмотреть классификатор, основанный на разделяющей гиперплоскости, которая не идеально разделяет два класса.
Это выполняется из соображений:
* Большей устойчивости к отдельным наблюдениям;
* Лучшей классификации по большинству тренировочных наблюдений;

Т.е. в худшем случае мы некорректно проклассифицируем несколько объектов тренировочной выборки для того, чтобы классификатор более точно классифицировал другие наблюдения.

**Классификатор на опорных вектора**, иногда называемый **классификатором с мягким зазором** делает в точности описанную выше процедуру. Вместо того, *чтобы искать возможный максимальный зазор для каждого наблюдения не только на корректной стороне от гиперплоскости, а также и на корректной стороне зазора, вместо это мы допускаем, что некоторые наблюдения будут лежать на некорректной стороне от зазора, или даже от гиперплоскости (Зазор мягкий, поскольку он может быть "нарушен" некоторыми объектами тренировочной выборки). Пример изображен на левой части рис. 6. Большинство наблюдений тренировочной выборки располагаются с правильной стороны от зазора. Однако, небольшое число наблюдений расположено с некорректной стороны зазора.*

Наблюдение может быть не только с не той стороны зазора, но и с не с той стороны гиперплоскости. По факту, когда не существует разделяющей гиперплоскости, такая ситуация неизбежна. Наблюдения на некорректной стороне гиперплоскости соответствуют объектам тренировочной выборки, которые неправильно классифицированы классификатором на опорных векторах. ### Особенности классификатора на опорных векторах

Классификатор на опорных векторах классифицирует контрольное наблюдение в зависимости от того, с какой стороны от гиперплоскости оно расположено.

Гиперплоскость выбирается так, чтобы разделить большинство тренировочных объектов на два класса, но может некорректно классифицировать некоторые объекты. Это представляет собой решение следующей оптимизационной проблемы:
$$ max_{\beta_0, \beta_1, \ldots, \beta_p, \epsilon_1, \ldots, \epsilon_n} M$$
$$ \mbox{ при условии, что } \sum_{j=1}^{p}\beta_j^{2}=1,$$
$$ y_i (\beta_0 + \beta_1 x_{i1} + \beta_2 x_{i2} + \ldots + \beta_{p} x_{ip}) \geq M(1-\epsilon_i), \forall i =1,\ldots, n.$$
$$\epsilon_i \geq 0, \sum_{i=1}^{n}\epsilon_i \leq C$$


$C$ - неотрицательный гиперпараметр. 

$M$ - ширина зазора. Мы пытаемся найти максимум этой величины. $\epsilon_1, \ldots, \epsilon_n$ называются **фиктивными переменными**, которые позволяют отдельным наблюдениям тренировочной выборки находится на неправильной стороне относительна границы зазора или относительно гиперплоскости.

После решения оптимизационной задачи, контрольное наблюдение $x^{*}$ классифицируется как и ранее, простым определением с какой стороны гиперплоскости оно расположено.
Т.е. классификация проводится, основываясь на знаке: $f(x^{*}) = \beta_0 + \beta_1 x_1^{*} + \beta_2 x_2^{*} + \ldots + \beta_p x_p^{*}$.

Итак, фиктивные переменные $\epsilon_i$ говорят нам о том, где расположено $i$-е наблюдение относительно гиперплоскости и относительно границы зазора. 
Так, если $\epsilon_i = 0$, тогда $i$-е наблюдение лежит на правильной стороне относительно границы зазора. Если $\epsilon_i > 0$, тогда $i$-е наблюдение лежит нна неправильной стороне относительно границы зазора, и говорят, что *$i$-е наблюдение нарушило эту границу* В случае, если $\epsilon_i > 1$, тогда наблюдение лежит на неправильной стороне относительно гиперплоскости.

**Роль гиперпараметра $C$**
$C$ определяет сумму $\epsilon_i$ и тем самым задает допустимое число нарушений границы зазора (или гиперплоскости) и их выраженность.

$C$ можно рассматривать как **бюджет для количества раз, когда зазор может быть нарушен $n$ наблюдениями**.
В случае $C = 0$, то возможности для нарушений границы зазора нет, в связи с чем $\epsilon_1 = \epsilon_2 = \ldots = \epsilon_n = 0$, т.е. задача вырождается в задачу оптимизации для классификатора с максимальным зазором.

Если $C>0$, то не больше чем $C$ наблюдений могут быть на неправильной стороне относительно гиперплоскости, поскольку если наблюдение лежит на некорректной стороне относительно гиперплоскости тогда $\epsilon_i>1$, а мы требуем, чтобы $\sum_{i=1}^{n}\epsilon_i \leq C$.

По мере роста бюджета $C$ допустимая степень нарушений границ зазора возрастает, и поэтому зазор расширяется. И наоборот, по мере уменьшения $C$, допустимая степень нарушений границ зазора уменьшается и зазор становится уже.
На рис. 7 можно увидеть иллюстрацию.
![](img/svm/Theory7.png "Рис. 7.")

* На рисунке иллюстрирован классификатор на базе опорных векторов, который настроен с помощью четырех разных значений параметра $C$. Наибольшее значение $C$ используется в верхней левой части рисунка, а самое малое значение $C$ - в верхней правой части рисунка. При больших значениях $C$ допустимая степень нарушений границ зазора высока, и поэтому зазор достаточно широким. По мере уменьшения $C$ эта степень уменьшается и как следствие зазор сужается.*

На практике, параметр $C$ подбирается посредством кросс-валидации. $С$ контролирует баланс смещения-дисперсии статистической техники.

Рассмотренная оптимизационная задача обладает интересным свойством: оказывается, что только те наблюдения тренировочной выборки, которые или лежат на зазоре, или нарушают зазор будут влиять на гиперплоскость, а следовательно, и на получившийся классификатор.

Другими словами, наблюдение, которое находится строго на правильной стороне относительно границы зазора не влияет на классификатор на опорных векторах. Изменение положения этого наблюдения никак не изменило бы классификатор, при условии, что это наблюдение по-прежнему лежит на правильной стороне от границы зазора. Наблюдения, лежащие непосредственно на границах зазора, или на неправильной (для своего класса) стороне относительно этих границ, называются**опорными векторами (support vector)**. Эти наблюдения имеют влияние на классификатор на основе опорных векторов.


Тот факт, что решающее правило классификатора на базе опорных векторов основано только на небольшом подмножестве тренировочной выборки (опорных векторах) означает, что он является достаточно устойчивым к наблюдениям, которые достаточно удалены от гиперплоскости. Это свойство отличает классификатор на базе опорных векторов от множества других классификаторов.


## Машины опорных векторов (Support Vector Machines)

### Классификация с использованием нелинейных решающих границ

Классификатор на опорных векторах является естественным подходом к классификации в случае бинарной классификации, если граница между классами линейная. Однако, на практике, мы нередко встречаемся с нелинейными границами класса.

Например, рассмотрим данные на рисунке 8 слева. Ясно, что классификатор на опорных векторах, или любой линейный классификатор будет выполняться на этих данных достаточно плохо. И действительно, решающая граница, построенная классификатором бесполезна (рис. 8, справа).
![](img/svm/Theory8.png "Рис. 8.")

Теперь вспомним линейную регрессию. Линейная регрессия работает очень плохо, если есть нелинейные взаимосвязи между предикторами и целевой переменной. В этом случае, происходит расширение пространства атрибутов (feature space) с использованием функций от предикторов (например, квадратичных и кубических, чтобы добавить нелинейность). 

В случае классификатора на базе опорных векторов, мы можем решить проблему возможных нелинейных границ между классами похожим способом, т.е. расширив пространство атрибутов используя квадратичные, кубические или полиномы высших порядков как функции от предикторов.

Например, вместо использования классификатора на базе опорных векторов с использованием $p$ атрибутов:
$$X_1, X_2, \ldots, X_p,$$
мы можем использовать классификатор на базе опорных векторов с использованием $2p$ атрибутов:
$$X_1, X_1^2, X_2, X_2^2, \ldots, X_p, X_p^2.$$

Тогда, оптимизационная задача изменится следующим образом:
$$\max_{\beta_0, \beta_{11},\beta_{12},  \ldots, \beta_{p1},\beta_{p2}, \epsilon_1, \ldots, \epsilon_n} M$$
$$ y_i (\beta_0 + \sum_{j=1}^{p}\beta_{j1}x_{ij}+\sum_{j=1}^{p} \beta_{j2}x{ij}^2) \geq M(1-\epsilon_i), \forall i =1,\ldots, n.$$
$$ \mbox{ при условии, что } \sum_{j=1}^{p} \sum_{k=1}^{2}\beta_{jk}^{2}=1,$$
$$\epsilon_i \geq 0, \sum_{i=1}^{n}\epsilon_i \leq C$$

Но почему это приводит к нелинейной решающей границе? В расширеном пространстве атрибутов решающая граница, полученная из этой оптимизационной задачи в действительности является линейной. Однако, в исходном пространстве атрибутов решающая граница имеет вид $q(x)=0$, где $q$ - является полиномом второй степени, и её решения тоже являются нелинейными.

Мы могли бы расширить пространство атрибутов также при помощи полиномов более высоких порядков и с использованием переменных вида $x_{j}x_{j'}$ для $j\neq j'$, отражающих взаимодействие между предикторами.

В качестве альтернативы можно было бы рассмотреть и другие функции предикторов, отличные от полиномов. Нетрудно увидеть, что есть множество способов расширить пространство предикторов и что без соблюдения осторожности мы в итоге могли бы получить огромное число переменных, что сделало бы вычисления невозможными.

Теперь рассмотрим машину опорных векторов, которая позволяет расширить пространство предикторов для классификатора на опорных векторах так, чтобы вычисления оставались эффективными.

### Машина опорных векторов

Машина опорных векторов (SVM) представляет собой более общую версию классификатора на опорных векторах, которую получают путем расширения пространства предикторов особым образом с помощью *ядерных функций*.

Основная идея: мы можем расширить пространство предикторов (атрибутов) для учета нелинейной формы границы между классами. Использование ядерных функций есть не что иное, как эффективный способ реализации этой идеи.

Решение оптимизационной проблемы:
$$ max_{\beta_0, \beta_1, \ldots, \beta_p, \epsilon_1, \ldots, \epsilon_n} M$$
$$ \mbox{ при условии, что } \sum_{j=1}^{p}\beta_j^{2}=1,$$
$$ y_i (\beta_0 + \beta_1 x_{i1} + \beta_2 x_{i2} + \ldots + \beta_{p} x_{ip}) \geq M(1-\epsilon_i), \forall i =1,\ldots, n.$$
$$\epsilon_i \geq 0, \sum_{i=1}^{n}\epsilon_i \leq C$$

для классификатора на основе опорных векторов базируется на использовании только *скалярных произведений* наблюдений, а не самих исходных наблюдений.

Скалярное произведение векторов $a$ и $b$ вычисляется как $\langle a, b \rangle = \sum_{i=1}^{r}a_i b_i$. Тогда скалярное произведение двух наблюдений $x_i, x_{i'}$ определяется как:
$$\langle x_i, x_{i'} \rangle = \sum_{j=1}^{p}x_{ij}x_{i'j}$$


**Линейный классификатор на опорных векторах** можно представить как: 
$$f(x) = \beta_0 + \sum_{i=1}^{n}\alpha_i\langle x, x_i \rangle,$$
где имеется $n$ параметров $\alpha_i, i=1,\ldots,n$, по одному на каждое обучающее наблюдение.

Всё что нам требуется для получения оценок параметров $\alpha_1, \ldots, \alpha_n$ и $\beta_0$ - это $C^n_2$ скалярных произведений $\langle x_i, x_{i'} \rangle$ между всеми парами обучающих наблюдений.

Для нахождения функции $f(x)$ необходимо вычислить скалярное произведение между новым наблюдением $x$ р каждым из обучающих наблюдений $x_i$. Оказывается, однако, что при решении этой проблемы $\alpha_i$ принимают ненулевые значения **только для опорных векторов**, т.е. если обучающее наблюдение не является опорным вектором, то его параметр $\alpha_i = 0$.

Таким образом, если $S$ представляет собой множество индексных номеров этих опорных точек, то мы можем представить любое решение для функции $f(x) = \beta_0 + \sum_{i=1}^{n}\alpha_i\langle x, x_i \rangle$ в виде уравнения:
$$f(x) = \beta_0 + \sum_{i\in S}\alpha_i\langle x, x_i \rangle,$$
которое обычно содержит намного меньше членов, чем $f(x) = \beta_0 + \sum_{i=1}^{n}\alpha_i\langle x, x_i \rangle.$

А теперь представьите, что каждый раз, когда скалярное произведение $\langle x_i, x_{i'} \rangle = \sum_{j=1}^{p}x_{ij}x_{i'j}$ появляется в выражении  $f(x) = \beta_0 + \sum_{i=1}^{n}\alpha_i\langle x, x_i \rangle$, т.е. при нахождении классификатора на опорных векторах мы заменяем его на сл. обобщенную функцию скалярного произведения: 
$$K(x_i, x_{i'}),$$
где $K$ - это некоторая функция, которую мы будем называть **ядром**. Ядро представляет собой функцию, которая количественно выражает сходство между двумя наблюдениями.

Например, мы могли бы просто взять функцию:
$$K(x_i, x_{i'}) = \sum_{j=1}^{p}x_{ij}x_{i'j},$$
которая дала бы обычный классификатор на опорных векторах. Это уравнение известно как **линейное ядро**, поскольку классификатор на опорных векторах является линейным по предикторам. По сути, линейное ядро количественно выражает сходство между парой наблюдений в виде стандартного коэффициента корреляции Пирсона.

Однако, можно было бы выбрать и другую форму для ядра $K(x_i, x_{i'})$, например:
$$K(x_i, x_{i'}) = (1+\sum_{j=1}^{p}x_{ij}x_{i'j})^d.$$
Данное уравнение известно как **полиномиальное ядро степени $d$**, где $d$ - положительное целое число. Использование такого ядра с $d>1$ вместо стандартного линейного ядра в алгоритме классификации приводит к гораздо более гибкой решающей границе. 

В сущности, происходит построение классификатора на опорных векторах в пространстве большей размерности, которое, в отличие от пространства исходных переменных, включает также полиномы степени $d$. При объединении классификатора на основе опорных векторов с нелинейным ядром, получают классификатор, который называется **машиной опорных векторов**.


В этом случае (нелинейная) функция принимает форму:
$$ f(x) = \beta_0 + \sum_{i\in S} \alpha_i K(x, x_i).$$

Полиномиальное ядро - один из возможных примеров нелинейных ядер, однако существует и множество других альтернатив. Например, популярным является *радиальное ядро* сл. вида:
$$ K(x_i, x_{i'}) = \exp^{(-\gamma \sum_{j=1}^{p}(x_{ij} - x_{i'j})^2)}.$$

Где $\gamma$ - просто положительная константа. 

Как именно работает радиальное ядро? Если некоторое контрольное наблюдение $x^{*} = (x_1^{*}, \ldots, x_p^{*})^T$ находится далеко от обучающего наблюдения $x_i$ с точки зрения евклидова расстояния, то величина $\sum_{j=1}^{p}(x_{ij} - x_{i'j})^2$ будет большой, в связи с чем значение $ K(x_i, x_{i'}) = \exp^{(-\gamma \sum_{j=1}^{p}(x_{ij} - x_{i'j})^2)}$ будет очень низким.

Это значит, что $x_i$ е будет играть никакой роли в определении $f(x^{*})$ в $ f(x) = \beta_0 + \sum_{i\in S} \alpha_i K(x, x_i)$.

Метка класса, предсказываемая для контрольного наблюдения $x^{*}$ зависит от знака $f(x^{*})$. Другими словами, обучающие наблюдения, находящиеся далеко от $x^{*}$ практически не будут играть никакой роли в предсказании класса $x^{*}$. Это значит также, что **радиальное ядро проявляет очень локальное поведение в том смысле, что на метку класса контрольного наблюдения влияют только близко расположенные к нему наблюдения**.

В чем состоит преимущество ядер в сравнении с простым расширением пространства признаков при помощи функции исходных признаков?

Главное преимущество состоит в вычислительном характере и состоит оно в том, что при использовании ядер требуется вычислить $K(x_i, x_i')$ только для всех $C^n_2$ уникальных пар $i, i'$. Это можно сделать без непосредственной работы с расширенным пространством признаков. Данное обстоятельство является важным, поскольку во многих случаях применения SVM расширенное пространство признаков настолько велико, что вычисления становятся практически невозможными. Для некоотрых ядер, таких как радиальное ядро, пространство признаков является *неявным* и обладает бесконечной размерностью, в связи с чем мы в любом случае не смогли бы выполнить в нем никаких вычислений.


# Ядра специального вида: tree kernel


Граф - это тип структурированных данных в форме вершин, или узлов (vertex, nodes) и их взаимосвязей, описанных в виде ребер (edges).
Вершины и ребра в графах могут иметь ряд атрибутов, которые могут быть или категориальными, или даже более сложными.

Сегодня, значительные объемы данных доступны в форме графов или сетей (World Wide Web, социальные сети, семантические сети, биологические сети и т.д.).

Дерево - специальный тип графа, и он естественным образом представляет ряд типов данных. Анализ деревьев - важная область в анализе данных и компьютерных науках.

Древесные ядра (tree kernel) - это метод для сравнения деревьев друг с другом, позволяющий числнно выразить степень их схожести или различий. Это важный процесс для множества современных приложений классификации и анализа текстовых данных.

В задаче классификации, классы уже известны и модель классификации строится из тренировочной выборки, в которой корректный класс уже дан.
Задача кластеризации пытается идентифицировать классы, когда они неизвестны, группируя данные в категории, базируясь на некоторой мере их схожести.

Кластеризация может быть скомбинирована с теорией графов для идентификации групп схожих сетей деревьев. Древесные структуры данных используются для моделирования объектов разных областей. Например, в NLP деревья разбора (parse tree) моделируются упорядоченными и помеченными деревьями.

Полуструктурированные данные, такие как HTML и XML документы также могут моделироваться упорядоченными, помеченными деревьями.

#### Проклятие размерности

Большинство алгоритмов требуют, чтобы данные были преобразованы в векторную форму, представляющую значения данных как атрибуты (features) в пространстве атрибутов (feature space), поэтому данные могут быть проанализированы в пространстве признаков с использованием методов линейной алгебры.
В структурированных или полуструктурированных данных (подобных деревьям), размерность результирующих векторов (т.е. число атрибутов в пространстве атрибутов) может быть достаточно большим, поскольку пространство атрибутов должно сохранять информацию о структуре.

Это может быть сильным недостатком, поскольку много алгоритмов классификации не способны эффективно масштабироваться с ростом размерности входных данных. Другими словами, их точность классификации может снижаться с ростом размерности входа. Эта проблема известна как **проклятие размерности**.

Чтобы понять причину, стоящую за этим падением производительности, рассмотрим пространство $X$ размерности $d$. Положим, что $X$ содержит множество равномерно распределенных точек. Если размерность пространства $X$ растет, необходимое число точек, для сохранения одинаковой плотности должно **расти экспоненциально!** Другими словами, чем больше размерность пространства тем больше наши данные будут "разряженными". В общем, разряженные наборы данных не дают достаточно информации для построения хорошего классификатора, поскольку корреляции между элементами данных слишком слабы для того, чтобы алгоритм мог их детектировать.

![](https://uploads.toptal.io/blog/image/92116/toptal-blog-image-1455213842360-81859975464e7075ee7cdd1d1a5bb7c7.jpg)

Векторная репрезентация структурных данных уместна, когда информация об области может быть эффективно использована для выбора управляемого набора функций. Когда такой информации недостаточно, желательно использовать техники, способные напрямую обрабатывать структурные данные, без выполнения операций в векторном пространстве.





#### Ядерные методы (kernel methods)

Ядерные методы позволяют избежать конвертирования данных в векторную форму. Они требуют только измерения насколько каждая пара объектов в наборе данных близка друг к другу.

Эта мера называется **ядром** и функция для её определения называется **ядерной функцией**. Ядерные методы ищут линейные отношения в пространстве атрибутов. Функционально, это эквивалентно взятию скалярного произведения (dot product) двух объектов в пространстве признаков. Однако, ядерные методы избегают прямого оперирования в пространстве признаков, поскольку можно показать, что возможна замена на положительно полуопределенную симметричную функцию, которая может оперировать входными данными в исходном пространстве.

Преимуществом ядерных функций является то, что огромное пространство атрибутов может быть проанализировано с вычислительной сложностью, не зависящей от размера пространства атрибутов, а от сложности ядерной функции, а это означает, что ядерные методы не страдают от проклятия размерности.

Если мы рассмотрим конечные набора данных из $N$ объектов, мы можем представить их схожесть с помощью генерирования ядерной матрицы (kernel matrix) размером $N \times N$. Эта матрица независит от размера отдельных объектов. Это свойство полезно, когда мы имеем небольшой набор объектов с огромным пространством атрибутов.

В общем, ядерные методы базируются на разных ответах на вопрос репрезентации данных. Вместо отображения множества точек в пространство атрибутов, данные представляются через попарное сравнение в матрице $K$, и анализ может проводится над ядерной матрицей.

![](https://uploads.toptal.io/blog/image/92117/toptal-blog-image-1455213875679-7e56ac5662c68956bf34e6bf9a76cd9e.jpg)

Множество методов анализа данных могут использовать ядра. Для классификации данных, которые могут быть представлены деревьями с помощью ядерных методов, может, например, использоваться алгоритм SVM, который требует определить валидную (симметричную положительно определенную) ядерную функцию $k : T\times T \to R$, также называемую *tree kernel*. 

По соображениям практичности, можно требовать вычисления за полиномиальное время (т.е. не слищком долго) и должна быть способна распозновать изоморфные графы. Такие древесные ядра (tree kernel) называются *complete tree kernel*.

**String Kernel**

Сначала рассмотрим введение в string kernel, что поможет нам ввести еще одн метод, который базируется на преобразовании деревьев в строки.

Пусть $num_y(s)$ - число вхождений подстроки $y$ в строку $s$, а за $\|s\|$ обозначим длину строки.
Тогда, строковое ядро (string kernel) может быть определено следующим образом:

$$K_{string}(S_1, S_2) = \sum_{s \in F} num_s(S_1) num_s(S_2) w_s$$

где $F$ - множество подстрок, которые появляются в $S_1$ и $S_2$, параметр $w_s$ является весовым параметром (например, для выделения важных подстрок).

Это ядро даёт большее значение парам строк, которые содержат больше общих подстрок.

Древесные ядра, основанные на конвертировании деревьев в строки

Итак, рассмотренное строковое ядро можно использовать для построения древесного ядра. Идея, лежащая в основе этого состоит в преобразовании двух ядер в две строки, единообразным способом, который кодирует древесную структуру, а затем применяет к ним строковое ядро.

Итак, два дерева преобразуются в две строки следующим образом:

Пусть 
* $T$ обозначает целевое дерево
* $label(n_s)$ - метка строки для узла $n_s$ в дереве $T$
* $tag(n_s)$ обозначает строковое представление поддерева $T$, с корнем в $n_s$.

Итак, если $n_{root}$ корень дерева $T$, то $tag(n_{root})$ - строковое представление всего дерева $T$.

Пусть $string(T) = tag(n_{root})$ обозначает строковое представление дерева $T$. Рекурсивно применяются шаги снизу вверх для получения $string(T)$:
* Если вершина $n_s$ - листовая, то $tag(n_s) = $ "[" + $label(n_s)$ + "]" (где + представляет собой конкатенацию строк);
* Если вершина $n_s$ не листовая и имеет $c$ потомков $n_1, n_2, \ldots, n_c$, то $tag(n_1), tag(n_2), \ldots, tag(n_c)$ перечисляется в том же порядке, для получения $tag(n_{1*}), tag(n_{2*}), \ldots, tag(n_{c*})$ и тогда:

$tag(n_s)=$"["+$label(n_s)+tag(n_{1*})+tag(n_{2*})+\ldots+tag(n_{c*})$+"]".

Сл. картинка иллюстрирует пример этого преобразования из дерева в его строковое представление. Результатом является строка, начинающаяся с открытия квадратной скобки "[" и заканчивающаяся её закрытием "]", с каждой вложенной парой скобок-разделителей, соответствующих поддеревьям соответствующей вершины.

![](https://uploads.toptal.io/blog/image/92118/toptal-blog-image-1455213894818-a8faff0c148aed09994140dcdbf631b5.jpg)


Ну а теперь можно применить соответствующее преобразование для пары деревьев $T_1$ и $T_2$, для получения двух строк $S_1$ и $S_2$. После этого, можно с легкостью применить строковое ядро, описанное выше.

Древесное ядро (tree kernel) для деревьев $T_1$ и $T_2$, имеющих строковое представление $S_1$ и $S_2$ может быть определено следующим образом:

$K_{tree}(T_1, T_2) = K_{string}(string(T_1), string(T_2)) = K_{string}(S_1, S_2) = \sum_{s\in F}num_s (S_1) num_s (S_2) w_s$



В большинстве приложений, весовой параметр становится $w_{\|s\|}$, весом подстроки, основанной на её длине $\|s\|$. Типичными методами взвешивания для $w_{\|s\|}$ являются:
* Константное взвешивание ($w_{\|s\|} = 1$)
* $k$-спектральное взвешивание ($w_{\|s\|}=1$, если $\|s\|=k$ и $w_{\|s\|}=0$ в противном случае)
* Экспоненциальное взвешивание ($w_{\|s\|} = \lambda^{\|s\|}$, где $0\leq\lambda\leq 1$ - параметр затухания).

Для вычисления ядра, достаточно определить общие подстроки $F$, и подсчитать число их вхождений как в $S_1$, так и в $S_2$. 

**Техническая информация: ** Этот дополнительный шаг является достаточно хорошо изученной проблемой и решается за время $O(|S_1| + |S_2|)$ с использованием суффиксных деревьев (suffix tree) и суффиксных массивов (suffix arrays). Если мы допустим что максимальое число букв (битов, байтов, символов), необходимых для описания метки вершины константно, то длины конвертируемых деревьев $|S_1| = O(|T_1|)$ и $|S_2| = O(|T_2|)$. Поэтому, сложность вычисления ядерной функции составляет $O(|T_1|+|T_2|)$, т.е. линейная сложность.

#### Древесные ядра, основанные на подпутях (Tree Kernel based on Subpaths)

Древесные ядра, описанные выше используют "горизонтальный" подход (также называемый обходов в ширину), для конвертации деревьев в строки. 
В то время как этот подход является достаточно простым, это преобразование означает, что мы не можем оперировать с деревьями напрямую в их исходной форме.

Определим древесное ядро, оперирующие с вертикальными структурами в дереве, что позволит ядру работать с деревьями напрямую.

Участок пути от коря к одному из листьев называется подпутем (subpath), и множество подпутей (subpath set) - множество всех подпутей, включенных в дерево:

![](https://uploads.toptal.io/blog/image/92119/toptal-blog-image-1455213915536-b99c4879693851619c8945d80d220701.jpg)

Положим, что мы хотим определить древесное ядро (tree kernel) $K(T_1, T_2)$ между двумя деревьями $T_1$ и $T_2$. Используя множество подпутей (subpath set), мы можем определить это ядро следующим образом:
$K(T_1, T_2) = \sum_{p\in P} num_p(T_1) num_p (T_2)w_{|p|}$
где $num_p(T)$ - число раз, когда подпуть (subpath) появляется в дереве $T$, $|p|$ - число вершин в подпути $p$, и $P$ - множество всех подпутей (set of all subpath) в $T_1$ и $T_2$. $w_{|p|}$ - весовой коэффициент (похожий на то, что определено выше).

*Опциональный материал для интересующихся*
Ниже представлен **простой пример**  реализации ядра с использованием поиска в глубину. Хотя этот алгоритм выполняется за квадратичное время, существуют более эффективные алгоритмы, использующие суффиксные деревья или суффиксные массивы, или расширения базирующиеся на алгоритме quicksort, которые позволяют в среднем достигнуть логлинейного вреени выполнения: $O(|T_1|log|T_2|)$.

```python
subpath_track = {}

def generate_subpaths(path, l):
  if l == len(path):
    if tuple(path) not in subpath_track:
      subpath_track[tuple(path)] = 1
    else:
      subpath_track[tuple(path)] += 1
  else:
    index = 0
    while l+index-1 < len(path):
      if tuple(path[index: l+index]) not in subpath_track:
        subpath_track[tuple(path[index: l+index])] = 1
      else:
        subpath_track[tuple(path[index: l+index])] += 1
      index += 1

    generate_subpaths(path, l+1)


def get_subpaths(graph, root, track, path):
  track[root] = True

  if graph.degree(root) == 1:
    generate_subpaths(path, 1)
  else:
    for node in graph.neighbors(root):
      if node not in track:
        get_subpaths(graph, node, track, path + [node, ])


def kernel_subpath(t1, t2, common_track):
  kernel_v = 0
  for p in subpath_track:
    kernel_v += common_track[t1][p]*common_track[t2][p]

  return kernel_v
```

В приведеном выше коде использовался весовой параметр $w_{|p|}=1$. Это приведет к тому, что все подпути рассматриваются как равновесные.
Однако, есть много случае, когда используется $k$-спектральное взвешивание, или некоторые динамически назначаемые весовые значения.



#### Пример: Категоризация веб-сайтов

В качестве примера можно рассмотреть одно из приложений классификации с использованием деревьев: категоризация веб-сайтов. Страницы с разных веб-сайтов могут быть довольно эффективно категоризированы с использованием деревьев, благодаря схожести в их структуре.

Как мы можем провести категоризацию? HTML документы имеют вложенную логическую структуру, которая очень напоминает деревья. Каждый документ содержит корневой элемент с дополнительными вложенными элементами. Вложенные элементы в HTML теге являются логически эквивалентными узлам-потомкам.

![](https://uploads.toptal.io/blog/image/92120/toptal-blog-image-1455213933056-de3e24225f5d48e1164833c82633d428.jpg)

Пример участка кода, иллюстрирующий преобразование html-кода в дерево:

```python
def html_to_dom_tree(root):
  node_id = 1
  q = deque()
  graph = nx.Graph()

  q.appendleft({'element': root, "root_id": node_id})
  while len(q):
    node = q.pop()

    if node and node['element'].name == "body":
      graph.add_node(node_id, element=node['element'].name)
      node_id += 1

    root_id = node['root_id']
    labels[root_id] = node['element'].name

    for t in node['element'].contents:
      if t and t.name:
        graph.add_node(node_id, element=t.name)
        graph.add_edge(root_id, node_id)
        q.appendleft({"element": t, "root_id": node_id})
        node_id += 1

  return graph
```

Этот участок кода может создать структуру подобную той, что иллюстрирована на рисунке:
![](https://uploads.toptal.io/blog/image/92121/toptal-blog-image-1455213944957-2da62d281aa6b417b901c041794ae208.jpg)

Участок кода, приведенный выше использует ряд крайне полезных библиотек языка Python:
* NetworkX - для работы со сложными графовыми структурами
* Beutiful Soup - для трансфера данных с веб, парсинга веб страниц и их манипулирования.

Вызов 
    html_to_dom_tree(soup.find("body"))
вернет графовый объект NetworkX, с корнем в <body> элементе веб-страницы.

Пусть мы хотим найти группы в 1000 домашних веб-страниц. КОнвертируя каждую страницу в дерево, можно сравнивать их дрг с другом одним из ядер, рассмотренных выше. Определяя их схожесть друг с другом, мы можем найти, что, например, сайты интернет магазинов, новостные сайты, блоги, образовательные сайты с легкостью можно идентифицировать по их схожести друг с другом.
