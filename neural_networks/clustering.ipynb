```python
import numpy as np
import neurolab as nl
import pylab as pl

fig_size = pl.rcParams["figure.figsize"]
fig_size[0] = 20
fig_size[1] = 10
pl.rcParams["figure.figsize"] = fig_size
```

```python
n=3
center = np.array([[0.2, 0.2], [0.4, 0.4], [0.7, 0.3]])
```

```python
x = np.array([[0, 7], [4.7, 9.7], [4.6, 4.6], [1.3, 7.7], [4.4, 9.7], [4.6, 3.6], [0.4, 7.5], [4.3, 9.9], [5.6, 3.9], [0.2, 7], [4.8, 8.1], [5.1, 4.3], [0.3, 8], [4.3, 9.3], [4, 3.3], [0, 7.6], [4.2, 8.6], [5.2, 3.8], [0.1, 7.6], [4.2, 8.6], [5.2, 3.8], [0.1, 7.5], [4.3, 9.5], [4.5, 3.3]])
```

```python
net = nl.net.newc([[0.0, 10.0],[0.0, 10.0]], 3)
error = net.train(x, epochs=200, show=20)
```

```python
w = net.layers[0].np['w']
pl.plot(center[:, 1], 'yo', x[:,0], x[:,1], 'b+', w[:,0], w[:,1], 'ro')
pl.legend(['real centers','train samples', 'train centers'])
pl.show()
```
