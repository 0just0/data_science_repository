```python
import numpy as np
import neurolab as nl
import pylab as pl
import matplotlib
matplotlib.style.use('ggplot')
fig_size = pl.rcParams["figure.figsize"]
fig_size[0] = 20
fig_size[1] = 10
pl.rcParams["figure.figsize"] = fig_size
```

```python
#Количество классов
n = 4

x = np.array([[1., 5.], [2., 6.], [4., 4.], [5., 6.], [6., 3.], [6., 5.], [2., 3.], [3., 3.], [3., 1.], [7., 1.], [8., 2.], [8., 4.]])
```

```python
net = nl.net.newc([[0.0, 10.0],[0.0, 10.0]], 4)
error = net.train(x, epochs=200, show=20)
```

```python
w = net.layers[0].np['w'] #центры
pl.plot(x[:,0], x[:,1], 'b+', w[:,0], w[:,1], 'ro')
pl.legend(['train samples','centers'])
pl.show()

```
