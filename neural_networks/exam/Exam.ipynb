```python
import numpy as np
import neurolab as nl
import pylab as pl
import matplotlib
matplotlib.style.use('ggplot')
fig_size = pl.rcParams["figure.figsize"]
fig_size[0] = 20
fig_size[1] = 10
pl.rcParams["figure.figsize"] = fig_size
```

```python
#Количество классов
n = 4

x = np.array([[0.5, 2.], [1., -0.8], [2., -2.], [3.5, -2.2], [4.2, -1.2], [5.5, 1.6], [6., 2.7], [7, 3.9], [8.5, 4], [10, 4]])
```

```python
net = nl.net.newc([[0.0, 10.0],[-3., 10.0]], 4)
error = net.train(x, epochs=200, show=20)
```

```python
w = net.layers[0].np['w'] #центры
pl.plot(x[:,0], x[:,1], 'b+', w[:,0], w[:,1], 'ro')
pl.legend(['train samples','centers'])
pl.show()

```
