```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import neurolab as nl
matplotlib.style.use('ggplot')
fig_size = plt.rcParams["figure.figsize"]
fig_size[0] = 20
fig_size[1] = 10
plt.rcParams["figure.figsize"] = fig_size
```

```python
xtrain = np.array([[1., 1.], [4, 2.], [1.5, -0.5], [0.5, -2.], [1.5, -2.5], [3., -2.5], [4., -2.], [4.5, 0]]) 
```

```python
d = np.array([[1], [1], [1], [1], [-1], [-1], [-1], [-1]])
d = d.reshape(len(d),1)
d
```

```python
plt.plot(xtrain[:, 0], xtrain[:, 1], 'b+')
plt.plot([0., 5.], [-5., 5.], 'b')
plt.show()
```

```python
MyFirstNet = nl.net.newff([[0, 5], [-5, 5]], [3,1])
err = MyFirstNet.train(xtrain, d, show=15)
y = MyFirstNet.sim(xtrain) #Даем обучающий пример
print(y)
```

```python
plt.scatter([0, -4], [5, 2], 'b')
for i in range(len(xtrain)):
    if y[i] > 0:
        plt.plot(xtrain[i, 0], xtrain[i, 1], 'ro')
    if y[i] < 0:
        plt.plot(xtrain[i, 0], xtrain[i, 1], 'go')

plt.show()
```
