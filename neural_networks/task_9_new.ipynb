```python
import numpy as np
import neurolab as nl
import matplotlib
import matplotlib.pyplot as plt

from rbn import newpnn_linear
```

```python
matplotlib.style.use('ggplot')
fig_size = plt.rcParams["figure.figsize"]
fig_size[0] = 20
fig_size[1] = 10
plt.rcParams["figure.figsize"] = fig_size
```

```python
n=100
x = np.linspace(0, 5, n)

y = np.sin(x) * x

size = len(x)

inp = x.reshape(size,1)
tar = y.reshape(size,1)
```

```python
net = newpnn_linear(nl.tool.minmax(inp), 100, 1, 1000)
net.layers[0].np['w'] = inp
d = np.transpose(tar)
net.layers[1].np['w'] = d
```

```python
out = net.sim(inp)
```

```python
plt.plot(x, y, '-')
plt.plot(x, out, 'bo')
plt.show()
```
