```python
import neurolab as nl
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
matplotlib.style.use('ggplot')
fig_size = plt.rcParams["figure.figsize"]
fig_size[0] = 20
fig_size[1] = 10
plt.rcParams["figure.figsize"] = fig_size
```

```python
n=20
x = np.random.uniform(0, 5, (n, 2))

for i in range(n):
    if x[i, 1] - x[i, 0]**2 + 4*x[i, 0] - 5 > 0:
        final_y[i] = 1
    elif x[i, 1] - x[i, 0]**2 + 4*x[i, 0] - 5 < 0:
        final_y[i] = -1
final_y = final_y.reshape(n, 1)
final_y
```

```python
net = nl.net.newff([[0, 5], [0, 5]],[5, 1])

error = net.train(x, final_y, epochs=500, show=100, goal=0.02)

# Simulate network
out = net.sim(x)
out
```

```python
for i in range(n):
    if -1.1 < out[i] < -0.9:
        plt.plot(x[i, 0], x[i, 1], 'go')
    if 0.9 < out[i] < 1.1:
        plt.plot(x[i,0], x[i, 1], 'bo')
```

```python
x1 = np.linspace(0, 5, n)
b = x1[:]**2 - 4*x1[:] + 5

plt.plot(x1[:], b)

plt.show()
```
