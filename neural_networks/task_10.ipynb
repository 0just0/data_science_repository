```python
import numpy as np
import neurolab as nl
import matplotlib
import matplotlib.pyplot as plt

from rbn import Radial_Basis
```

```python
matplotlib.style.use('ggplot')
fig_size = plt.rcParams["figure.figsize"]
fig_size[0] = 20
fig_size[1] = 10
plt.rcParams["figure.figsize"] = fig_size
```

```python
n=20
x = np.linspace(0, 5, n)

y = np.sin(x) * x

size = len(x)

inp = x.reshape(size,1)
tar = y.reshape(size,1)
```

```python
net = nl.net.newff([[0, 5]],[5, 1], [Radial_Basis(), nl.net.trans.PureLin()])

error = net.train(inp, tar, epochs=500, show=100, goal=0.02)

# Simulate network
out = net.sim(inp)
```

```python
plt.plot(x, y, '-')
plt.plot(x, out, 'bo')
plt.show()
```
