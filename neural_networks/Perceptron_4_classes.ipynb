```python
import neurolab as nl
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
matplotlib.style.use('ggplot')
fig_size = plt.rcParams["figure.figsize"]
fig_size[0] = 20
fig_size[1] = 10
plt.rcParams["figure.figsize"] = fig_size
```

```python
n=200
x = np.random.uniform(0, 5, (n, 2))
#y_left_right = np.sign(x[:,0] + x[:, 1] - 5).reshape(n, 1)
#y_up_down = np.sign(x[:, 1] - x[:, 0]).reshape(n, 1)
#final_y = np.array([y_left_right, y_up_down]).reshape([n, 2])
final_y = np.empty([n, 2])

for i in range(n):
    if x[i,0] + x[i, 1] - 5 > 0:
        if x[i,1] - x[i, 0] > 0:
            final_y[i] = [1, 1]
        elif x[i,1] - x[i, 0] < 0:
            final_y[i] = [1, -1]
    elif x[i,0] + x[i, 1] - 5 < 0:
        if x[i,1] - x[i, 0] > 0:
            final_y[i] = [-1, 1]
        elif x[i,1] - x[i, 0] < 0:
            final_y[i] = [-1, -1]
final_y
```

```python
net = nl.net.newp([[0, 5],[0, 5]], 2, nl.net.trans.HardLims())
err_left_right = net.train(x, final_y, epochs=100, show=10)
a = net.sim(x)
a
```

```python
for i in range(n):
    if a[i, 0] == 1 and a[i, 1] == 1:
        plt.plot(x[i, 0], x[i, 1], 'bo')
    elif a[i, 0] == -1 and a[i, 1] == -1:
        plt.plot(x[i, 0], x[i, 1], 'ro')
    elif a[i, 0] == 1 and a[i, 1] == -1:
        plt.plot(x[i, 0], x[i, 1], 'b+')
    elif a[i, 0] == -1 and a[i, 1] == 1:
        plt.plot(x[i, 0], x[i, 1], 'r+')
```
