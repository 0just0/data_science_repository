```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df1=pd.read_csv('interactions_tiny.csv')
test=df1.iloc[-df1.shape[0]//3:]
train=df1.iloc[:-df1.shape[0]//3]
subm=test.groupby('item').count()#[['user']]
def f1(a):
    return a//2
subm.apply(f1).to_csv('test.csv', header=None)
subm
```

```python
def my_join(l):
    return (l-min(df1['timestamp'])).values#-min(l)

df3=train.groupby('item')['timestamp'].apply(my_join)

sample=df3.loc[983438]
xy=plt.hist(sample, bins=10, normed=True)
x=xy[0]
y=xy[1]
df3
```

```python
from numpy.polynomial import Polynomial

p = Polynomial.fit(x, y, 10)
plt.plot(x, y, '.',*p.linspace(n=100, domain=[0,16000]))
```
